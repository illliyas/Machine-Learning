```
import yfinance as yf
import pandas as pd
```

```
stocks=['ARE&M.NS',
'BAJAJ-AUTO.NS',
'DIVISLAB.NS',
'DRREDDY.NS',
'EXIDEIND.NS',
'INFY.NS',
'NATCOPHARM.NS',
'RELIANCE.NS',
'SUNPHARMA.NS',
'TATACHEM.NS',
'TATAMOTORS.NS',
'TATASTEEL.NS',
'TCS.NS',
'TECHM.NS',
'WIPRO.NS',
'ZYDUSLIFE.NS',
'CIPLA.NS',
'ABBOTINDIA.NS',
'HINDUNILVR.NS',
'NESTLEIND.NS',
'TVSMOTOR.NS',
'FINPIPE.NS',
'TATAPOWER.NS',
'GRAUWEIL.BO',
        'ACI.NS',
        'SHYAMMETL.NS',
        'NOCIL.NS',
        'AMIORG.NS',
        'KCP.NS',
        'SANSERA.NS',
        'ROLEXRINGS.NS',
        'AMBUJACEM.NS',
        'BORORENEW.NS'
]
```

```
cols = ['exchange','symbol','marketCap','currentPrice','trailingPE','trailingEps','profitMargins','ebitda',
'debtToEquity',
'trailingAnnualDividendYield',
'profitMargins',
'priceToBook',
'financialCurrency',
'earningsQuarterlyGrowth']
stocks_list=[]
for s in stocks:
  print(s)
  dic={}
  stock = yf.Ticker(s)
  for c in cols:
    try:
      dic[c]=stock.info[c]
    except:
      dic[c]=None
  try:
    hist = stock.history(period="1mo")
    hist = hist.sort_index(ascending=False)
    dic['DIFF_30_DAYS']=(hist.iloc[0]['Close']-hist.iloc[-1]['Close'])/hist.iloc[-1]['Close']
    dic['DIFF_7_DAYS']=(hist.iloc[0]['Close']-hist.iloc[5]['Close'])/hist.iloc[5]['Close']
    dic.update(dict(enumerate(map(str, stock.quarterly_income_stmt.loc['Net Income'].values))))
  except:
    None
  stocks_list.append(dic)
```

```
df = pd.DataFrame(stocks_list)
```
