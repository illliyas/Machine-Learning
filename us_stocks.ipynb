```
import yfinance as yf
import pandas as pd
```

```
stocks=['AAPL','MSFT','GOOGL',
'PG',
'MRK',
'KO',
'ABT',
'AMAT',
'PCAR',
'HPQ',
'ON',
'LLY',
'HON',
'TEL',
'KHC',
'SKX',
'GIS',
'TM',
'V',
'MCD',
'HMC']
```

```
cols = ['exchange','symbol','marketCap','currentPrice','trailingPE','trailingEps','profitMargins','ebitda',
'debtToEquity',
'trailingAnnualDividendYield',
'profitMargins',
'priceToBook',
'financialCurrency',
'earningsQuarterlyGrowth'
]
stocks_list=[]
for s in stocks:
  print(s)
  dic={}
  stock = yf.Ticker(s)
  stock_info = stock.get_info()
  for c in cols:
    try:
      dic[c]=stock_info[c]
    except:
      None
  try:
    hist = stock.history(period="1mo")
    hist = hist.sort_index(ascending=False)
    dic['DIFF_30_DAYS']=(hist.iloc[0]['Close']-hist.iloc[-1]['Close'])/hist.iloc[-1]['Close']
    dic['DIFF_7_DAYS']=(hist.iloc[0]['Close']-hist.iloc[5]['Close'])/hist.iloc[5]['Close']
    dic.update(dict(enumerate(map(str, stock.quarterly_income_stmt.loc['Net Income'].values))))
  except:
    None
  stocks_list.append(dic)
```

```
df = pd.DataFrame(stocks_list)
```

```
df[[0, 1,2,3,4]] = df[[0, 1,2,3,4]].apply(pd.to_numeric)
```

```
df['Q_PRE_Y_Q_VAR'] = (df[0]-df[4])/df[4]
df['Q_PRE_Q_VAR'] = (df[0]-df[1])/df[1]
```

```
df[((df['Q_PRE_Y_Q_VAR']>0) & (df['trailingPE']<=28))].sort_values('DIFF_30_DAYS')[['symbol','trailingPE','debtToEquity','DIFF_7_DAYS','DIFF_30_DAYS','Q_PRE_Y_Q_VAR','Q_PRE_Q_VAR']]
```
